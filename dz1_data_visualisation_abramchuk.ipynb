```python
countries = []
years = []
cigarettes_per_day = []

with open("smoking.csv", "r", encoding="utf-8") as csvfile:
    data1 = csv.reader(csvfile)

    print(data1)
    for raw in data1:
        countries.append(raw[0])
        years.append(raw[1])
        cigarettes_per_day.append(raw[2])

countries = np.array(countries[1::])

years = np.array(years[1::])

cigarettes_per_day = cigarettes_per_day[1::]
cigarettes_per_day = np.array([float(i) for i in cigarettes_per_day])
```

```python
counter = 0
count = 0

for i in range(len(countries)-1):
    if countries[i] == countries[i+1]:
        counter += 1
    else:
        print(count, "->", counter)
        count = counter
```

```python
plt.figure(figsize=[32, 16])
plt.subplot(221)
plt.bar(years[0:32],    cigarettes_per_day[0:32])
plt.xlabel('years')
plt.ylabel('cigarettes per day')


plt.subplot(222)
plt.bar(years[32:64],    cigarettes_per_day[32:64])
plt.xlabel('years')
plt.ylabel('cigarettes per day')


plt.subplot(223)
plt.bar(years[64:96],    cigarettes_per_day[64:96])
plt.xlabel('years')
plt.ylabel('cigarettes per day')


plt.subplot(224)
plt.bar(years[96:128],    cigarettes_per_day[96:128])
plt.xlabel('years')
plt.ylabel('cigarettes per day')

plt.show()
```
