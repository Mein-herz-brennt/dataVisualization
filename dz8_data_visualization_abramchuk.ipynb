```python
import pandas as pd
import csv
import matplotlib.pyplot as plt
import numpy as np

plt.style.use("seaborn-v0_8")
data = pd.read_csv("smoking.csv")
data.head()
```

```python
data_smokers = data[:33]
smokers_female = data_smokers["Data.Smokers.Female"]
smokers_male = data_smokers["Data.Smokers.Male"]
data_time = data_smokers["Year"]
data_total = data_smokers["Data.Smokers.Total"]
np_data_array = np.array([[smokers_male[0], smokers_female[0], data_total[0]], [smokers_male[16], smokers_female[16], data_total[16]], [smokers_male[32], smokers_female[32], data_total[32]]])
np_data_array1 = np.array([data["Data.Smokers.Total"][32], data["Data.Smokers.Female"][32], data["Data.Smokers.Male"][32]])
print(np_data_array, np_data_array1)
```

```python
a,  b, c = [plt.cm.Greens, plt.cm.Reds, plt.cm.Blues]
outer_colors = [a(.6), b(.6), c(.6)]
inner_colors = [a(.5), a(.4), b(.5), b(.4), c(.5), c(.4)]
group_names = ["male", "female", "total"]
size = 0.3
plt.figure()
plt.pie(np_data_array.sum(axis=1), radius=1, colors=outer_colors, labels=group_names, wedgeprops=dict(width=size, edgecolor="w"))
plt.pie(np_data_array.flatten(), radius=1-size, colors=inner_colors, wedgeprops=dict(width=size, edgecolor="w"))
plt.show()
```

```python
import squarify
plt.figure()
labels = ["Data.Smokers.Total", "Data.Smokers.Female", "Data.Smokers.Male"]
a, b = [plt.cm.Reds, plt.cm.Blues]
colors_ = [a(.4), a(.3), a(.2)]
plt.rc('font', size=14)
squarify.plot(sizes=np_data_array1, label=labels, color=colors_, alpha=0.7)
plt.axis('on')
plt.show()


```
