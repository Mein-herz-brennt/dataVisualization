```python
import pandas as pd
import csv
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns


plt.style.use("seaborn-v0_8")
data = pd.read_csv("unemployment-rate.csv")
data.head()
```

```python
sns.set(font_scale=0.8)
sns.lineplot(x="Year", y="Unemployment, total (% of total labor force) (modeled ILO estimate)", data=data[:200], hue="Entity")

countries = data["Entity"].unique()[:6]

for j in countries:
    y_ann = data[data['Entity']==j][data[data['Entity']==j]["Year"]==2021]["Unemployment, total (% of total labor force) (modeled ILO estimate)"]
    plt.annotate(j, xy=(2021, y_ann), xycoords="data", xytext=(2021, y_ann), textcoords="data")

```

```python
sns.lineplot(x="Year", y="Unemployment, total (% of total labor force) (modeled ILO estimate)", data=data[:200], hue="Entity")
sns.lineplot(x="Year", y="Unemployment, total (% of total labor force) (modeled ILO estimate)", data=data[:29], hue="Entity")
countries = data["Entity"].unique()[:6]

j = "Afghanistan"
y_ann = data[data['Entity']==j][data[data['Entity']==j]["Year"]==2021]["Unemployment, total (% of total labor force) (modeled ILO estimate)"]
plt.annotate(j, xy=(2021, y_ann), xycoords="data", xytext=(2021, y_ann), textcoords="data")
plt.legend(loc="upper right")
```
