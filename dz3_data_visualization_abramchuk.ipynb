```python
import pandas as pd
import csv
import matplotlib.pyplot as plt
import numpy as np

plt.style.use("seaborn-v0_8")
```

```python
data = pd.read_csv("smoking.csv")
data.head()
```

```python
data_smokers = data[:33]
data_smokers
```

```python
data_total = data_smokers["Data.Smokers.Total"]
data_total
```

```python
data_time = data_smokers["Year"]
data_time
```

```python
smokers_male = data_smokers["Data.Smokers.Male"]
smokers_male
```

```python
smokers_female = data_smokers["Data.Smokers.Female"]
smokers_female
```

```python
all_total = sum(data_total)
all_total
```

```python

plt.figure()
# plt.plot(smokers_male, data_time, label="male data by time")
# plt.plot(smokers_female, data_time, label="female data by time")
# plt.plot(data_total, data_time, label="Total smokers in year")
plt.stackplot(data_time, [data_total, smokers_male, smokers_female], colors=['purple', 'yellow', "green"], labels=["male", "female", "total"])
plt.legend()
plt.show()
```

```python
figure, axis = plt.subplots(2, 2)

axis[0,0].stackplot(data_time[:10], [smokers_male[:10], smokers_female[:10], data_total[:10]], colors=['purple', 'yellow', "green"], labels=["male smokers start", "female smokers start", "total smokers start"])
# axis[0,0].legend()
axis[0,0].set_title("start")

axis[0,1].stackplot(data_time[10:21], [smokers_male[10:21], smokers_female[10:21], data_total[10:21]], colors=['purple', 'yellow', "green"], labels=["male smokers start", "female smokers start", "total smokers start"])
axis[0,1].set_title("middle")

axis[1,0].stackplot(data_time[21:32], [smokers_male[21:32], smokers_female[21:32], data_total[21:32]], colors=['purple', 'yellow', "green"], labels=["male smokers start", "female smokers start", "total smokers start"])
axis[1,0].set_title("end")



```

```python
figure, axis = plt.subplots(2, 2)

axis[0,0].bar(data_time[:10], smokers_male[:10], color='purple')
axis[0,0].bar(data_time[:10], smokers_female[:10], color='green', bottom=smokers_male[:10])
axis[0,0].bar(data_time[:10], data_total[:10], color='blue', bottom=(smokers_male[:10]+smokers_female[:10]))
# axis[0,0].legend()
axis[0,0].set_title("start")

axis[0,1].bar(data_time[10:21], smokers_male[10:21], color='purple')
axis[0,1].bar(data_time[10:21], smokers_female[10:21], color='green', bottom=smokers_male[10:21])
axis[0,1].bar(data_time[10:21], data_total[10:21], color='blue', bottom=(smokers_male[10:21]+smokers_female[10:21]))
axis[0,1].set_title("middle")

axis[1,0].bar(data_time[21:32], smokers_male[21:32], color='purple')
axis[1,0].bar(data_time[21:32], smokers_female[21:32], color='green', bottom=smokers_male[21:32])
axis[1,0].bar(data_time[21:32], data_total[21:32], color='blue', bottom=(smokers_male[21:32]+smokers_female[21:32]))
axis[1,0].set_title("end")
```

```python
plt.figure()
years = [data_time[0], data_time[16], data_time[32]]
y1 = [smokers_male[0], smokers_male[16], smokers_male[32]]
y1 = np.array(y1)
y2 = [smokers_female[0], smokers_female[16], smokers_female[32]]
y3 = [data_total[0], data_total[16], data_total[32]]
plt.bar(years, y1, color='purple', width=5)
plt.bar(years, y2, color='green', bottom=y1, width=5)
plt.bar(years, y3, color='blue', bottom=(y1+y2), width=5)
```
