```python
import pandas as pd
import csv
import matplotlib.pyplot as plt
import numpy as np

plt.style.use("seaborn-v0_8")
data = pd.read_csv("smoking.csv")
data.head()
```

```python
data_smokers = data[:33]
smokers_female = data_smokers["Data.Smokers.Female"]
smokers_male = data_smokers["Data.Smokers.Male"]
data_time = data_smokers["Year"]
data_total = data_smokers["Data.Smokers.Total"]
```

```python
plt.figure()
plt.stackplot(data_time, [data_total, smokers_male, smokers_female], colors=['purple', 'yellow', "green"], labels=["male", "female", "total"])
plt.legend()
plt.show()
```

```python
from matplotlib import animation
fig = plt.figure()

plt.tick_params(axis='x',which='both',bottom=False,top=False,labelbottom=False)
ax = plt.axes()
#our masrterlist of artist objects
ims =[]
for n in range(2,len(data_time)-1):
    x = data_time[:n]
    y = [smokers_male[:n], smokers_female[:n], data_total[:n]]
    #each artist object is a stackplot figure
    im = ax.stackplot(x,y, labels=["male", "female", "total"],
             colors=['yellow','b','grey'])
    ims.append(im)

ani = animation.ArtistAnimation(fig, ims, interval=2, blit=True,repeat_delay=2)
ani.save("dz10.gif", writer="ffmpeg", fps=30)
# plt.show()
```
