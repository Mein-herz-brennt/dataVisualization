### DATA

https://www.kaggle.com/datasets/ajaypalsinghlo/world-happiness-report-2023/

```python
import pandas as pd
import csv
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns


plt.style.use("seaborn-v0_8")
data = pd.read_csv("WHR2023.csv")
data.head()
```

137 unique values
137 рядків та 19 стовбців

```python
plt.rc('axes', titlesize=4)
plt.rc('xtick', labelsize=4)
plt.rc('ytick', labelsize=4)

figure, axis = plt.subplots(3, 6)
axists = []
for j in range(3):
        for i in range(6):
            axists.append(axis[j, i])


n = 0

for key, value in data.items():
    if key != "Country name":
        axists[n].hist(value, histtype="bar", bins=40)
        axists[n].set_title(key)
        n += 1

plt.show()
```

### Нормалізована

```python
plt.rc('axes', titlesize=4)
plt.rc('xtick', labelsize=4)
plt.rc('ytick', labelsize=4)
figure, axis = plt.subplots(3, 6)
axists = []
for j in range(3):
        for i in range(6):
            axists.append(axis[j, i])


n = 0

for key, value in data.items():
    if key != "Country name":
        axists[n].hist(value, density=True, histtype="bar", bins=40)
        axists[n].set_title(key)
        n += 1

plt.show()
```
